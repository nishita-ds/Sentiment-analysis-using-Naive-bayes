# Text Cleaning
## 1.Sample 10000 rows
## 2.Remove html tags
## 3.Remove special characters
## 4.Convert every thing to lower case
## 5.Removing stop words
## 6.Stemming


```python
movie=movie.sample(10000)
```

```python
movie.shape
```

```python
movie.info()
```

```python
movie['sentiment'].replace({'positive':1,'negative':0},inplace=True)
```

```python
movie.head()
```

```python
import re
clean=re.compile('<.*?>')
re.sub(clean,'',movie.iloc[2].review)
```

```python
# Function to clean html tags
def clean_html(text):
    clean=re.compile('<.*?>')
    return re.sub(clean,'',text)
```

```python
movie['review']=movie['review'].apply(clean_html)
```

```python
# Converting everything to lowercase
def convert_lower(text):
    return text.lower()
```

```python
movie['review']=movie['review'].apply(convert_lower)
```

```python
movie.head()
```

```python
# Function to remove special characters
def remove_special(text):
    x=''

    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x
```

```python
movie['review']=movie['review'].apply(remove_special)
```

```python
#Remove the stop words
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
# Remove the stop words
def remove_stopwords(text):
    x=[]
    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    y=x[:]
    x.clear
    return y
```

```python
movie['review']=movie['review'].apply(remove_stopwords)
```

```python
movie
```

```python
# Stemming
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
y=[]
def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z
```

```python
movie['review']=movie['review'].apply(stem_words)
```

```python
# Join back
def join_back(list_input):
    return " ".join(list_input)
```

```python
movie['review']=movie['review'].apply(join_back)
```

```python
movie
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000)
```

```python
X=cv.fit_transform(movie['review']).toarray()
```

```python
X.shape
```

```python
y=movie.iloc[:,-1].values
```

```python
y.shape
```

```python
# splitting data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
x_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
clf1=GaussianNB()
clf2=MultinomialNB()
clf3=BernoulliNB()
```

```python
clf1.fit(x_train,y_train)
clf2.fit(x_train,y_train)
clf3.fit(x_train,y_train)
```

```python
y_pred1=clf1.predict(x_test)
y_pred2=clf2.predict(x_test)
y_pred3=clf3.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred1))
print(accuracy_score(y_test,y_pred2))
print(accuracy_score(y_test,y_pred3))
```
